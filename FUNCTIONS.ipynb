```python
# Write a Python program to print "Hello, World!
print("Hello, World!")
# Write a Python program that displays your name and age

# Define variables
name = "Your Name"  # Replace with your name
age = 20  # Replace with your age

# Display name and age
print("Name:", name)
print("Age:", age)
# Write code to print all the pre-defined keywords in Python using the keyword library
# Import the keyword library
import keyword

# Get a list of all Python keywords
keywords = keyword.kwlist

# Print the keywords
print("Python Keywords:")
for keyword in keywords:
    print(keyword)
# Write a program that checks if a given word is a Python keyword?
# Import the keyword library
import keyword

# Function to check if a word is a Python keyword
def is_python_keyword(word):
    return keyword.iskeyword(word)

# Get the word from the user
word = input("Enter a word: ")

# Check if the word is a Python keyword
if is_python_keyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is not a Python keyword.")
#  Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each?
# Create a list
my_list = [1, 2, 3, 4, 5]
print("Original List:", my_list)

# Attempt to change an element in the list
my_list[0] = 10
print("List after modification:", my_list)

# Create a tuple
my_tuple = (1, 2, 3, 4, 5)
print("\nOriginal Tuple:", my_tuple)

# Attempt to change an element in the tuple
try:
    my_tuple[0] = 10
except TypeError as e:
    print("Error:", e)
# Write a function to demonstrate the behavior of mutable and immutable arguments
def modify_args_immutable(x, y):
    # Attempt to modify immutable arguments
    x = 10  # This will not affect the original value
    y = "New String"  # This will not affect the original value
    print("Inside function - Immutable arguments:")
    print("x =", x)
    print("y =", y)

def modify_args_mutable(lst):
    # Attempt to modify mutable arguments
    lst.append(10)  # This will affect the original list
    lst[0] = 100  # This will affect the original list
    print("Inside function - Mutable arguments:")
    print("lst =", lst)

# Test with immutable arguments
x = 5
y = "Original String"
print("Before function call - Immutable arguments:")
print("x =", x)
print("y =", y)
modify_args_immutable(x, y)
print("After function call - Immutable arguments:")
print("x =", x)
print("y =", y)

# Test with mutable arguments
lst = [1, 2, 3]
print("\nBefore function call - Mutable arguments:")
print("lst =", lst)
modify_args_mutable(lst)
print("After function call - Mutable arguments:")
print("lst =", lst)
# Write a function to demonstrate the behavior of mutable and immutable arguments?
def modify_args(x, y, lst):
    """
    Demonstrates the behavior of mutable and immutable arguments.

    Args:
        x (int): Immutable argument
        y (str): Immutable argument
        lst (list): Mutable argument
    """
    # Attempt to modify immutable arguments
    x = 10  # This will not affect the original value
    y = "New String"  # This will not affect the original value

    # Attempt to modify mutable arguments
    lst.append(10)  # This will affect the original list
    lst[0] = 100  # This will affect the original list

    print("Inside function:")
    print("x =", x)
    print("y =", y)
    print("lst =", lst)

# Test the function
x = 5
y = "Original String"
lst = [1, 2, 3]

print("Before function call:")
print("x =", x)
print("y =", y)
print("lst =", lst)

modify_args(x, y, lst)

print("After function call:")
print("x =", x)
print("y =", y)
print("lst =", lst)

# Write a program to demonstrate the use of logical operators
def demonstrate_logical_operators():
    """
    Demonstrates the use of logical operators.
    """

    # Define variables
    x = 5
    y = 3

    # Logical AND operator (and)
    print("Logical AND operator (and):")
    print("x =", x)
    print("y =", y)
    print("(x > 3) and (y < 5):", (x > 3) and (y < 5))  # Expected output: True
    print("(x > 3) and (y > 5):", (x > 3) and (y > 5))  # Expected output: False

    # Logical OR operator (or)
    print("\nLogical OR operator (or):")
    print("x =", x)
    print("y =", y)
    print("(x > 3) or (y < 5):", (x > 3) or (y < 5))  # Expected output: True
    print("(x < 3) or (y < 5):", (x < 3) or (y < 5))  # Expected output: True

    # Logical NOT operator (not)
    print("\nLogical NOT operator (not):")
    print("x =", x)
    print("y =", y)
    print("not (x > 3):", not (x > 3))  # Expected output: False
    print("not (x < 3):", not (x < 3))  # Expected output: True

# Call the function
demonstrate_logical_operators()

#Write a Python program to convert user input from string to integer, float, and boolean types
def convert_input():
    """
    Converts user input from string to integer, float, and boolean types.
    """

    # Get user input
    user_input = input("Enter a value: ")

    # Convert to integer
    try:
        int_value = int(user_input)
        print(f"Integer value: {int_value}")
    except ValueError:
        print("Cannot convert to integer.")

    # Convert to float
    try:
        float_value = float(user_input)
        print(f"Float value: {float_value}")
    except ValueError:
        print("Cannot convert to float.")

    # Convert to boolean
    bool_value = user_input.lower() == "true"
    print(f"Boolean value: {bool_value}")

# Call the function
convert_input()

#Write code to demonstrate type casting with list elements
def demonstrate_type_casting():
    """
    Demonstrates type casting with list elements.
    """

    # Create a list with mixed data types
    my_list = [1, 2.5, "hello", True, None]

    # Print original list
    print("Original List:")
    print(my_list)

    # Type casting: Convert all elements to string
    str_list = [str(element) for element in my_list]
    print("\nList after converting all elements to string:")
    print(str_list)

    # Type casting: Convert all elements to integer (where possible)
    int_list = []
    for element in my_list:
        try:
            int_list.append(int(element))
        except (ValueError, TypeError):
            int_list.append(None)
    print("\nList after converting all elements to integer (where possible):")
    print(int_list)

    # Type casting: Convert all elements to float (where possible)
    float_list = []
    for element in my_list:
        try:
            float_list.append(float(element))
        except (ValueError, TypeError):
            float_list.append(None)
    print("\nList after converting all elements to float (where possible):")
    print(float_list)

# Call the function
demonstrate_type_casting()

#Write a program that checks if a number is positive, negative, or zero
def check_number(num):
    """
    Checks if a number is positive, negative, or zero.

    Args:
        num (int): The number to check.

    Returns:
        str: "Positive" if the number is positive, "Negative" if it's negative, and "Zero" if it's zero.
    """
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

  # Write a for loop to print numbers from 1 to 100
  # For loop to print numbers from 1 to 100
for i in range(1, 101):
    print(i)

 # Write a Python program to find the sum of all even numbers between 1 and 500
 # Initialize sum variable
total_sum = 0

# Iterate over even numbers between 1 and 500
for num in range(2, 501, 2):
    total_sum += num

# Print the result
print("The sum of all even numbers between 1 and 500 is:", total_sum)

# Write a program to reverse a string using a while loo
def reverse_string(input_str):
    """
    Reverses a string using a while loop.

    Args:
        input_str (str): The input string to be reversed.

    Returns:
        str: The reversed string.
    """
    reversed_str = ""
    index = len(input_str) - 1

    while index >= 0:
        reversed_str += input_str[index]
        index -= 1

    return reversed_str

 # Write a Python program to calculate the factorial of a number provided by the user using a while loop.
 def calculate_factorial(n):
    """
    Calculates the factorial of a number using a while loop.

    Args:
        n (int): The number for which the factorial is to be calculated.

    Returns:
        int: The factorial of the input number.
    """
    factorial = 1
    i = 1

    while i <= n:
        factorial *= i
        i += 1

    return factorial
```
